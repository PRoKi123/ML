```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig = plt.figure() #создали объект figure


#набор больших осей
axes_1 = fig.add_axes([0, 0, 1, 1]) #повторили размер объекта figure [x,y,размер, размер]
axes_1.set_xlim(0, 8) #предел по оси x
axes_1.set_ylim(0, 8000) #предел по оси y
axes_1.set_xlabel('A') #название оси x
axes_1.set_ylabel('B') #название оси y
axes_1.set_title('Возведение в степень 4') #название графика
axes_1.plot(a,b); #получили график


#набор маленьких осей
axes_2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes_2.set_xlim(1, 2) #предел по оси x
axes_2.set_ylim(0, 50) #предел по оси y
axes_2.set_xlabel('A') #название оси x
axes_2.set_ylabel('B') #название оси y
axes_2.set_title('Увеличенный фрагмент') #название графика
axes_2.plot(a,b); #получили второй график, который меньше первого


#если сделать координаты x,y = 1,1, то график будет за границами объекта figure
#также график будет за границами объекта, если ширину сделать последние 2 аргумента сделать больше 1
```

Параметры объекта figure

```python
fig = plt.figure(dpi=200, figsize=(2,2))
#dpi - добавляет пиксели и делает картинку четче
#figsize - размер объекта figure(в дюймах)


axes_1 = fig.add_axes([0, 0, 1, 1])
axes_1.plot(a,b)


fig.savefig('new_figure.png', bbox_inches='tight')
#сохранили график в формате png
#bbox_inches - команда для отображения осей на картинке
```
