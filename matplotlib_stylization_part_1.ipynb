```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])


axes.plot(x, x, label='X, X')
#построили график и использовали параметр label= для описания графика
axes.plot(x, x**2, label='X, X^2')
#построили график и использовали параметр label= для описания графика


axes.legend();
#вывели легенду с описаниями графиков
#есть параметр loc= для расположения легенды.
#loc='lower left' - для расположения легенды в левои нижнем углу
#loc=0 - для автоматического выбора наилучшего места для расположения легенды
#axes.legend(loc=(1.1, 0.5)) - сами задали координаты расположения легенды: сначала по горизонтали, затем по вертикали
```
