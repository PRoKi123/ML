Для функции groupby() мы должны выбрать категориальную колонку.
>
Категориальные колонки принимают дискретные(не непрерывные) значения.
>
Они могут быть и числовыми, например класс вагона в поезде или на корабле:
>
1-й класс, 2-й класс, 3-й класс.


Операция groupby() создает 'lazy' объект groupby, который не выполняет всю работу сразу, а ждет того момента, когда мы обратимся к этому объекту, и данные действительно понадобятся.



```python
df.groupby('model_year').mean(numeric_only=True)#['mpg'] если только для 1 колонки
#группировка по одной колонке
#работает с агрегирующими функциями.
#получили средние значения колонок по годам.
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True)
#группировка по 2 и более колонкам.
```

Работа с мульти индексами

```python
year_cyl = df.groupby(['model_year', 'cylinders']).mean(numeric_only=True)
```

```python
year_cyl.index.names
#вывели названия индексов
```

```python
year_cyl.index.levels
#вывели уровни индексов.
```

```python
year_cyl.loc[70]
#работа с внешним индексом.
#вывели все значения где model_year равен 70.
#model_year не в индексе.
```

```python
year_cyl.loc[[70,82]]
#работа с внешним индексом.
#вывели все значения где model_year равен 70 или 82.
#model_year остается в индексе.
```

```python
year_cyl.loc[(70,4)]
#получили одну строку со значениями 70-й год и 4 цилиндра.
```

Операция Cross-Section

year_cyl.xs(key,axis = 0 or 1, level)

```python
year_cyl.xs(key=70, level = 'model_year')
#вывели все строки со значением model_year = 70
```

key только одно значение

```python
year_cyl.xs(key=4, level='cylinders')
#вывели все строки с машинами, у которых 4 цилиндра
```

xs не показывает тот уровень индекса, по которому была фильтрация данных

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year', 'cylinders']).mean()
#вывели строки только с выбранными значениями цилиндра и затем отсортировали также нашли среднее значение столбцов
```

Меняем внешний и внутренний индекс местами

```python
year_cyl.swaplevel()
#поменяли внешний и внутренний индекс местами
```

Сортировка мультииндекса

```python
year_cyl.sort_index(level='model_year',ascending=False)
#сортировка внешнего индекса в порядке убывания
```

```python
year_cyl.sort_index(level='cylinders',ascending=False)
#сортировка внутреннего индекса по убыванию. Влияет на отображение внешнего индекса
```

Лучше сортировать по внешнему индексу, а если нужно по внутреннему, то сначала поменять уровни местами

```python
Применение разных функций агрегации для разных колонок
```

```python
df.head()
```

```python
df.agg(['std','mean'])
#применили несколько функций агрегации ко всем колонкам
#в будущих версиях может вылезти ошибка, что функцию применили к столбику, к которому их нет смысла применять
```

```python
df.agg(['std','mean'])['mpg']
#применили разные функции к одному столбцу
#в будущих версиях может вылезти ошибка, что функцию применили к столбику, к которому их нет смысла применять
```

```python
df.agg({'mpg' : ['max', 'min'], 'weight' : ['mean', 'std']})
#применили к конкретным колонкам конкретные функции
#значения, которые не планировалось вычислять, заменены на NaN
```
