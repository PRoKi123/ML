```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Scatter plot - диаграма рассеивания
>
Показывает связь между двумя непрерывными переменными
>
Непрерывная переменная - это числовая переменная, которая может принимать любые промежуточные значения
>
Примеры: возраст, высота, зарплата, температура
>
В непрерывной переменной всегда найдется какое-то третье значение
>
Это сильно отличается от категориальных(дискретных) перменных, где каждое значение - отдельная категория
>
Примеры: цвета, имена, формы и т.д

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(12,4), dpi=200) #можно менять размер, dpi диаграммы


sns.scatterplot(x='salary', y='sales', data=df, hue = 'level of education', palette = 'Dark2') #построили диаграмму


#Дополнительные параметры
#hue = 'level of education' - расскрасили точки в зависимости от значений в колонке (указываем категориальную переменную)
#palette = 'Dark2' - указали другую палитру цветов
#size = 'salary' - размер точки в зависимости от значений в колонке
#s = 200 - размер всех точек
#alpha = 0.2 - прозрачность точек (от 0 до 1)
#style = 'level of education' - различные маркеры в зависимости от значений в колонке (указываем категориальную переменную)


plt.savefig('my_plot.png') #сохранили график
```
