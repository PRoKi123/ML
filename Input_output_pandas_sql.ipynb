```python
temp_db = create_engine('sqlite:///:memory:')
#создали временную базу данных в оперативной памяти компьютера
```

```python
df = pd.DataFrame(data=np.random.randint(low=0, high=100, size=(4,4)), columns=['a', 'b', 'c', 'd'])
```

```python
df
```

```python
df.to_sql(name='new_table', con=temp_db, if_exists='replace')
#сохранили базу данных в sql формате.
#параметр if_exist='fail', 'replace', 'append' если таблица с таким именем уже существует
```

```python
new_df = pd.read_sql(sql='new_table', con=temp_db.connect())
new_df
#прочитали всю базу данных sql целиком
```

```python
result = pd.read_sql_query(sql=text('SELECT A,C FROM new_table'), con=temp_db.connect())
result
#сделали sql запрос и вывели только нужную часть базы данных sql
```
