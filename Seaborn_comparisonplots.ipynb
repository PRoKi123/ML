```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

Графики сравнения - это двумерные версии тех графиков, которые мы уже изучили

jointplot
>
С помощью него мы можем взять гистограммы для каждой переменной графика scatterplot, чтобы уточнить распределение данных по каждой переменной
>
Кроме scatterplot можно также использовать hexplot(hexagon plot) или двумерный KDE plot


```python
sns.jointplot(x='math score', y='reading score', data=df, hue='gender')
#построили график jointplot
#kind='hex' - вместо стандартного графика scatterplot взяли hexagonplot
#alpha=0.2 - изменили прозрачность точек
#kind='hist' - вместо стандартного графика scatterplot взяли гистрограмму
#kind='kde' - вместо стандартного графика scatterplot взяли KDE plot + fill=True - добавили оттенки на график
#hue='gender' - добавили распределение по полу
```

pairplot
>
Это быстрый способ сравнить между собой все числовые переменные в наборе данных
>
Автоматически создает гистограмму для каждой переменной, и сравнительные графики scatter plot для всех комбинаций перменных
>
Потребляет много CPU и оперативной памяти. Так что хорошая идея - сначала оставить только те колонки, которые нужно исследовать

```python
sns.pairplot(data=df, hue='gender', corner=True)
#построили график pairplot
#hue='gender' - добавили распределение по полу
#diag_kind='hist' - построили гистограммы на главной диагонали(по умолчанию график KDE)
#corner=True - убрали дубликаты графиков


'''
Сначала лучше сделать отбор коррелирующих переменных на графике pairplot,
а затем две интересующие нас переменные изобразить на jointplot
'''
```
