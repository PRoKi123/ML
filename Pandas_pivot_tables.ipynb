```python
import pandas as pd
import numpy as np
```

Сводные таблицы позволяют поменять структуру данных, преобразуя ячейки данных с помощью нового индекса.

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
```

```python
df
```

Пример: сколько лицензий купила компания Google?

```python
licenses = df[['Company', 'Product', 'Licenses']]
```

```python
pd.pivot(data=licenses, index='Company', columns='Product', values='Licenses')
#вывели сколько продуктов приобрела каждая компания
#если информации нет, будет значение NaN
```

Пример: найти суммарное количество лицензий и сумму продаж.

```python
pd.pivot_table(df,index='Company', aggfunc='sum', values=['Licenses', 'Sale Price'])
#получили сумму для числовых колонок по компаниям
```

Пример: группировка по менеджерам и контактам и узнаем сумму продаж.

```python
pd.pivot_table(df, index=['Account Manager', 'Contact'],
               columns=['Product'],
               values=['Sale Price'],
               aggfunc=['sum', 'mean'],
               fill_value=0,
               margins=True)
#Можно сортировать по 2 колонкам, расположенным в строках и 1 колонку как колонку
#fill_value - значение вместо NaN
#в aggfunc можно передать несколько функций в виде списка
#margins = True or False - позволяет получить общую сумму внизу датафрейма
```
