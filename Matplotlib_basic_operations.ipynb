```python
import matplotlib.pyplot as plt
import numpy as np
```

Самый простой способ задать график - метод plt.plot(x,y)
>
Он хорош для быстрой визуализации данных
>
p.s. Если скрипт с расширением .py, то для отображения графика нужна команда plt.show()

```python
x = np.arange(0,10)
y = 2 * x
```

```python
plt.plot(x,y) #построили график по коороднитам
plt.title('Название графика') #дали название графику. В конце знак ; чтобы не было текста над названием графика
plt.xlabel('ось X') #дали название оси X
plt.ylabel('ось Y') #дали название оси Y
plt.xlim(0,6) #задали лимит значений оси X и вывели только часть графика
plt.ylim(0,15);#задали лимит значений оси Y и вывели только часть графика
plt.savefig('My_first_plot.png') #сохранили график в формате png
#все команды должны быть в одной ячейке
```
