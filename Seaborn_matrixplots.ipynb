Матричные графики
>
-это визуальный аналог отображения сводной таблицы
>
Матричный график показывает все данные, визуализируя все числовые значения в dataframe
>
Замечание! Не каждый dataframe хорошо подходит для матричных графиков, например для тепловой карты(heatmap)
>
Типы графиков:
>
heatmap() - тепловая карта - изображает распределение значений по ячейкам с помощью цветов
>
clustermap() - карта кластеров - похожа на heatmap, но сначала выполняет иерархическую кластеризацию, объединяя данные в группы

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
df.head()
```

```python
df = df.set_index('Countries')
#сделали индексами значения колонки 'Countries'
```

График heatmap

```python
plt.figure(dpi=105)


sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidth=5, annot=True, cmap='viridis')
#построили график heatmap
#data=df.drop('Life expectancy' - в параметры data указали все колонки кроме 'Life expectancy'
#linewidth=5 - изменили расстояние между ячейками
#annot=True - указали параметр для того, чтобы видеть значения внутри ячеек
#cmap='viridis' - измениди палитру цветов
#centre=40 - изменили цетровку шкалы(лучше оставить по умолчанию)
```

График clustermap

```python
plt.figure(dpi=105)


sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidth=5, annot=True, cmap='viridis', col_cluster=False)
#построили график clustermap и сгруппировали похожие строки и столбцы друг с другом(объединили в кластеры)
#параметры такие же
#col_cluster=False - убрали группировку столбцов
```
