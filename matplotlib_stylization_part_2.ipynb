```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])


axes.plot(x,x, c='red', lw=2,ls='-.', marker='o',ms=20,markerfacecolor='orange', markeredgewidth=4, markeredgecolor='blue')
axes.plot(x,x+1,c='#851031', lw=2,ls='--',marker='+',ms=20,markerfacecolor='green', markeredgewidth=4, markeredgecolor='pink');
#параметр c='' отвечает за цвет линии графика
#также в параметр c='' можно указать код rgb цвета
#параметр lw= отвечает за ширину линии графика
#параметр ls= отвечает за вид линии графика(можно сделать пунктирной и т.д)
#параметр marker= отвечает за стиль точек на графике
#параметр ms= отвечает за размер маркера
#параметр markerfacecolor= отвечает за цвет центра маркера
#параметр markeredgewidth= отвечает за размер центра маркера
#параметр markeredgecolor= отвечает за цвет маркера


#lines = axes.plot(x,y)
#lines[0].set_dashes([3,1,3,3,5])
#задали свой стиль линии графика
#функция set_dashes([сколько точек нарисовать, сколько точек пропустить, сколько точек нарисовать, сколько точек пропустить])
```
