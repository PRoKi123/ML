Создание Series

```python
import numpy as np
import pandas as pd
```

```python
my_index = ["Usa", "Canda", "Mexico"]
my_data = [1776, 1867, 1821]
#сделал именованные индексы и данные в объект list
```

```python
my_ser = pd.Series(data=my_data, index = my_index)
my_ser
#Создали объект Series
```

```python
my_ser[0]
my_ser["Usa"]
#можно обращаться к данным по индексу или по именованному индексу
```

```python
ages = {"Sam": 5, "Frank": 10, "Spike": 7}
pd.Series(ages)
#создание объекта Series на основе словаря. Индексы уже именованные
```

Операции над Series

```python
q1 = {'Japan': 80, 'China': 450, 'India': 200, 'USA': 250}
q2 = {'Brazil': 100,'China': 500, 'India': 210,'USA': 260}
```

```python
sales_q1 = pd.Series(q1)
sales_q2 = pd.Series(q2)
```

```python
sales_q1.keys()
#посмотреть ключи объекта
```

```python
sales_q1 * 2 #все значения умножили на 2 (broadcast)
```

```python
sales_q1 + sales_q2 #сложили 2 объекта Series, но имеются значения NaN
```

```python
sales_q1.add(sales_q2, fill_value = 0 ) #сложили 2 объекта Series, заменили NaN на 0
```

Так же при создании pd.Series(тут можно указать dtype(опрдеелнный тип данных, который нам нужен))

Выбор строк по условию 2 способа

```python
#1 способ. Сумма счета больше 40
bool_series = df['total_bill'] > 40
df[bool_series]
```

```python
#2 способ. Сумма счета больше 40
df[df["total_bill"] > 40]
```

```python
df[df['sex'] == 'Male'] #только мужчины
```

```python
df[df['size'] > 3] #количество гостей больше 3
```

Выбор строк по нескольким условиям
>
and заменяется на &
>
or заменяется на |

```python
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')] #выбрали строки с суммой счета больше 30 и платили мужчины
```

```python
df[(df['day'] == 'Sat') | (df['day'] == 'Sun')] #выбрали чеки в субботу или воскресенье
```

Хотим добавить еще несколько дней
>
Есть два способа
>
1) Добавить условие в строку кода выше
>
2) Метод isin() - True, если есть в списке значений


```python
df[df['day'].isin(['Sat','Sun'])] #можем добавлять значения в список
```
