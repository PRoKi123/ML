```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribution plots - отображают одну непрерывную переменную, и помогают визуализировать атрибуты этой перменной, например среднее значение или дисперсию
>
Три основных вида графиков:
>
1)Rug Plot
>
2)Histogram
>
3)KDE Plot (Kernel Density Estimation)

Rugplot - график, на котором видны значения заданной колонки, ось y не используется
>
KDE - ядерная оценка плотности. Способ оценки функции распределения случайной величины. Является непрерывным графиком.


```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
#plt.figure(figsize=(5,8), dpi=200) также можно настраивать график


sns.rugplot(x='salary', data=df, height=0.5) #построили график rugplot
#на ось y не обращаем внимание
#height = 0.5 - высота столбиков (50% от высоты всего графика)
```

```python
sns.set(style='darkgrid') #сделали темный фон и белую сетку


sns.displot(x='salary', data=df, bins = 20, kde = True) #построили гистограмму
#bins = 20 - количество интервалов
#kde = True - добавили график kde
#rug = True - добавили график rugplot
```

```python
sns.histplot(x='salary', data=df) #также строит гистограмму, но в displot можно добавить график kde, rugplot
```

```python
sns.kdeplot(x='salary', data=df) #сделали график kde
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
sample_ages
```

```python
sns.rugplot(x='age', data=sample_ages, height=0.5)
```

```python
sns.displot(x='age', data=sample_ages, bins=30, kde=True)
```

```python
sns.kdeplot(x='age', data=sample_ages, clip=[0,100], bw_adjust=0.6, fill=True)
#clip=[0,100] - интвервал по оси x от 0 до 100
#bw_adjust=0.6 - сделали более детальынй график (больше делений, как bins)
#fill=True - закрасили область внутри графика
```
