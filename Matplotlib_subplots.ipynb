```python
import numpy as np
import matplotlib.pyplot as plt
```

С помощью этой функции можно создать несколько графиков рядом друг с другом.
>
Возвращает кортеж, в котором находятся объект figure и массив numpy с объектами axes.

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
#nrows - количество строк.
#ncols - количество колонок.
#figsize(4,2) - размер графиков.


axes[0][0].plot(x,y)
#нарисовали график в первой строке первой колонке(левом верхнем углу).
axes[0,1].plot(a,b)


#fig.subplots_adjust(wspace=0.9, hspace=0.9).
#можно самому задать расстояние между графиками.
#wspace - расстояние по ширине(горизнотали).
#hspace - расстояние по высоте(вертикали).


axes[1][0].set_ylabel('Y_label')
axes[1][1].set_title('TITLE_1.1')
#даем название определенных параметров определенных графиков.


plt.suptitle('Figure_Level_Title')
#дали название набору графиков.


#fig.set_figwidth(10)
#увеличили ширину всех графиков.


plt.tight_layout()
#подбирает размер между графиками. Указывать в самом конце.


fig.savefig('new_subplots.png', bbox_inches='tight')
#сохранили набор графиков в формате png.
```
