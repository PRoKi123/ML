### ЗАДАНИЕ: воссоздайте график Scatter Plot, показанный ниже.

**График scatterplot пытается показать зависимость между количеством дней на работе (days employed) и возрастом человека (DAYS_BIRTH) для всех, кроме безработных. Обратите внимание, что для этого графика Вам нужно будет сначала удалить безработных людей из набора данных. Также обратите внимание на значения по осям - значения были преобразованы так, чтобы они были положительными. Также при желании можете поменять параметры *alpha* и *linewidth*, поскольку здесь много точек накладываются друг на друга.**

<img src="task_one.jpg">


```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ


res = df.copy()


res = res[res['DAYS_EMPLOYED'] < 0]
res['DAYS_BIRTH'] = res['DAYS_BIRTH'] * -1
res['DAYS_EMPLOYED'] = res['DAYS_EMPLOYED'] * -1


plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(x='DAYS_BIRTH', y='DAYS_EMPLOYED', data=res, alpha=0.01, linewidth=0)
```

### ЗАДАНИЕ: воссоздайте приведённый ниже график Categorical:

<img src='catplot_solution.png'>

**Этот график отображает информацию только для *половины всех людей*, имеющих доход в этом наборе данных. Причём мы берём "нижнюю" половину - тех, кто зарабатывает меньше. Графики boxplot отображены для каждого значения колонки NAME_FAMILY_STATUS, показывая распределения их общего дохода. Раскраска hue сделана по колонке "FLAG_OWN_REALTY". Замечание: чтобы взять только нижнюю половину распределения дохода, Вам нужно будет преобразовать dataframe или взять только его часть *перед* тем, как рисовать график.**

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ


half_income = df.nsmallest(n=int(len(df)/2), columns='AMT_INCOME_TOTAL')


plt.figure(figsize=(10,6))
sns.boxplot(x='NAME_FAMILY_STATUS', y='AMT_INCOME_TOTAL', data=half_income, hue='FLAG_OWN_REALTY', width=0.9)
#параметр order для сортировки категорий по оси x


plt.legend(loc=(1.1, 0.5), title='FLAG_OWN_REALTY')
```

### ЗАДАНИЕ: воссоздайте приведённый ниже график Heat Map:

<img src='heatmap_solution.png'>

**Эта тепловая карта heatmap показывает корреляцию между колонками в наборе данных. Корреляцию можно посчитать с помощью метода .corr().  Обратите внимание, что колонка FLAG_MOBIL имеет NaN-корреляцию со всеми другими колонками - это видно в результатах метода .corr(). Поэтому Вам следует удалить колонку FLAG_MOBIL перед построением графика.**

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
res = df.copy()


res = res.drop(['FLAG_MOBIL'], axis = 1)


sns.heatmap(res.corr(), cmap='viridis')
#sns.heatmap(data=df.drop(['FLAG_MOBIL'], axis = 1))
```
