```python
import numpy as np
import pandas as pd
```

Тип NaN значений:
>
1) np.nan(будем использовать)
>
2) pd.NA
>
3) Для TIMESTAMP pd.NaT

С NaN не работают операции сравнения
>
np.nan in nan - для проверка NaN или нет - это сработает
>
== не использовать, не сработает

```python
df = pd.read_csv('movie_scores.csv')
```

```python
df
```

```python
df.isnull() #вернет True для каждой ячейки, где есть значение NaN
```

```python
df.notnull() #вернет True для кажой ячейки, где значение отличное от NaN
```

```python
df['pre_movie_score'].notnull() #получили объект Series с True для каждого значения столбца pre_movie_score, отличного от NaN
```

```python
df[df['pre_movie_score'].notnull()]
#получили объект DataFrame с True для каждой строки, где значение pre_movie_score отличное от NaN
```

```python
df[(df['pre_movie_score'].isnull()) & (df['first_name'].notnull())]
#вывели строки, где в колонке pre_movie_score значение равно NaN , и где в колонке first_name значение отличное от NaN
```

Работа с отсутствующими данными

```python
#1)оставляем данные без изменений
df
```

```python
#удаление данных
df.dropna() #удалили все строки, где есть хоть 1 значение NaN
```

```python
df.dropna(thresh=1)
#также удалили строки со значением NaN, но оставили те строки, где есть хоятбы одно значимое значение
```

```python
df.dropna(axis=1)
#уже работаем с колонками. Удалили все столбцы, где есть хотябы 1 значение NaN
```

```python
df.dropna(subset=['last_name'])
#удалили строки, где значение NaN встречается в колонке last_name
```

```python
#3) замена отсутсвтвующих значений
df.fillna('Новое значение')
#вместо значения NaN будет НОвое значение
```

```python
df['pre_movie_score'].fillna(0)
#заменили все значения NaN в колонке pre_movie_score на значение 0
```

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
#заменили значения NaN срденим значением колонки, среднее значение формируется из значимых значений
```

```python
df.fillna(df.mean(numeric_only=True))
#заменили значения NaN на среднее значение во всех числовых колонках
```

```python
airline_tix = {'first':100,'business':np.nan,'economy-plus':50,'economy':30}
```

```python
ser = pd.Series(airline_tix)
ser
```
