```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

Seaborn grid
>
-это способ построения набора графиков рядом друг с другом. В основе лежит subplots() из matplotlib, с помощью которого автоматически создается набор графиков для категориальной переменной
>
Указывать количество колонок и строк не нужно, вместо этого мы указываем название колонки, и seaborn автоматически создает набор графиков(subplot)
>
Многие встроенные методы seaborn внутри себя используют seaborn grid
>
Если вызывать его напрямую, то мы получим возможность настроить различные параметры графиков

График catplot
>
-показывает статистическое распределение двух переменных

```python
sns.catplot(x='gender', y='math score', data=df, kind='box', row='test preparation course', col='lunch')
#построили график catplot
#kind='box' - указали какие график строить
#row='test preparation course' - на отдельных строках разделили графики для разных значений колонки 'test preparation course'
#col='lunch' - на отдельных столбиках разделили графики для разных значений колонки 'lunch'
```

```python
g = sns.PairGrid(data=df, hue = 'gender')
#построили набор осей для всех числовых переменных
#hue='gender' - добавили распределение по полу для всех графиков


g = g.map_upper(sns.scatterplot)
#в верхней части матрицы построили графики scatterplot


g = g.map_diag(sns.histplot)
#на главной диагонали построили гистограммы


g = g.map_lower(sns.kdeplot)
#в нижней части матрицы построили графики KDE
```
